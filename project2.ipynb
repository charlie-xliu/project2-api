## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed

URL = 'https://en.wikipedia.org/wiki/Phineas_and_Ferb'
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")

for data in soup(['style', 'script']):
  data.decompose()

print(' '.join(soup.stripped_strings))
```

```python
soup.get_text()
```

```python
#download an image with beautifulsoup and save it in this repository
img_tags=soup.find_all('img')
img_url = 'https:' + img_tags[10]['src']
```

```python
img=requests.get(img_url)

with open('image.jpg','wb') as f:
    f.write(img.content)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread('image.jpg')
imgplot = plt.imshow(img)
plt.show()
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links
links=soup.find_all("a")

link_url=[]
for i in links:

    #issues with some links not having href, we just use try except to skip it
    try:
        link_url.append(i['href'])
    except:
        continue
link_url
```

```python
link_url=[i if 'http' in i else 'https://en.wikipedia.org/'+i for i in link_url]
link_url
```

```python
for i in range(10):
  thing = link_url[i]
  print(thing[:100])
```

## Part 3: Downloading scripts

```python
scripts=pd.read_csv('pudding_data.csv')
```

```python
scripts
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags
def getChar(link):
  try:  
    page = requests.get(link)
    soup = BeautifulSoup(page.content, "html.parser")
    for data in soup(['style', 'script']):
      data.decompose()
    return ' '.join(soup.stripped_strings)[:1000]
  except:
    return None

scripts['output'] = scripts['link'].apply(getChar)
scripts
```

```python
#add a new column to the df with the text downloaded
#save this new dataframe as "pudding_texts.csv"

scripts.to_csv('pudding_texts.csv', index = False, escapechar='\\')
```

#### Browse the documentation at https://developer.themoviedb.org/reference/intro/getting-started. Create an account to authenticate

```python
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 
api_key = '712b05da78d28f9b4789ddfeaa7e7975'

url = f"https://api.themoviedb.org/3/movie/now_playing?api_key={api_key}"

# Sending the request
response = requests.get(url)
data = response.json()

df = pd.DataFrame(data['results'], columns=['title', 'release_date', 'vote_average', 'overview', 'original_language'])
df
```

```python
data['results']
```

```python
#download the movie posters for 10 of these movies and save them to this repository
import os

poster_folder = 'posters'
os.makedirs(poster_folder, exist_ok=True)

for movie in data['results']:
    poster_path = movie.get('poster_path')
    poster_url = f"https://image.tmdb.org/t/p/w500{poster_path}"

    poster_name = movie.get('title').replace(" ", "_") + '_poster.jpg'
    poster_path_local = os.path.join(poster_folder, poster_name)

    response = requests.get(poster_url)

    with open(poster_path_local, "wb") as file:
        file.write(response.content)
```
